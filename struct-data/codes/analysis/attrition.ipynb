```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
```

```python
df = pd.read_pickle('../../output/data/final/original_extended_final.pkl')
df = df[df['AGE'].le(47) & df['SURVEY_YEAR'].ge(1979)]
```

```python
df['CENSORED_NEXT_PERIOD'] = 0
last_year = df.groupby('IDENTIFIER').last().SURVEY_YEAR
cond = list(zip(last_year.index, last_year))
df.loc[(cond), 'CENSORED_NEXT_PERIOD'] = 1
df['CENSORED_NEXT_PERIOD'] = df['CENSORED_NEXT_PERIOD'].astype(int)
df['SURVEY_YEAR'] = df['SURVEY_YEAR'].astype(int)
```

```python
late_df = df[df['AGE'].le(46) & df['SURVEY_YEAR'].lt(2011) & df['SURVEY_YEAR'].gt(1994) & (df['SURVEY_YEAR'] % 2).eq(1)]
```

```python
school_age_brackets = [[16, 17], list(range(18, 21))]

work_age_brackets = [list(range(18, 21)), list(range(21, 26)), 
                       list(range(26, 31)), list(range(31, 36)), list(range(36, 41)), list(range(41, 48))]
```

```python

for bracket in work_age_brackets:
    
    df['employment_' + str(bracket[0]) + '_' + str(bracket[-1])] = (df['CHOICE'].isin(['blue_collar', 'white_collar', 'military']) 
                                                                                  & df['AGE'].isin(bracket)).astype(int)
        
    
    for choice in ['blue_collar', 'white_collar', 'military']:
        
        df[choice + '_' + str(bracket[0]) + '_' + str(bracket[-1])] = (df['CHOICE'].eq(choice) 
                                                                       & df['AGE'].isin(bracket)).astype(int)
        
    
```

```python
for bracket in school_age_brackets:
    
    df['school_' + str(bracket[0]) + '_' + str(bracket[-1])] = (df['CHOICE'].eq('schooling') 
                                                                & df['AGE'].isin(bracket)).astype(int)
  
```

```python
formula = 'CENSORED_NEXT_PERIOD ~ employment_31_35 + employment_36_40 + employment_41_47 + C(AGE) + C(SURVEY_YEAR)'

mod = smf.probit(formula, df)
res = mod.fit()
print(res.summary())
```
